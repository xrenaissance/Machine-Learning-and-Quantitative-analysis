# Bool Selection
```
and/or in python only deal with one boolean value against another boolean value

&/| in python deal with a series bool values
```

````python
df[(df['W'] > 0) & (df['Y'] > 1)]
````

````python
(df['W'] > 0) & (df['X'] > 1)
````

````python
(df['W'] > 0) | (df['X'] > 1)
````
