## Filtering Data

A large part of working with DataFrames is the ability to quickly filter out data based on conditions. Spark DataFrames are built on top of the Spark SQL platform, which means that is you already know SQL, you can quickly and easily grab that data using SQL commands, or using the DataFram methods.

```python
df.describe().show()
```

```python
df.show(3)
```

```python
# Using SQL
df.filter('Close < 500').show()
```

```python
# Using SQL with .select
df.filter('Close < 500').select(['Open', 'Close']).show()
```

```python
df.filter(df['Close'] < 500).select('Volume').show()
```

Using normal python comparison operators is another way to do this, they will look very similar to SQL operators, except you need to make sure you are calling the entire column within the dataframe, using the format: df["column name"]

```python
# Make sure to add in the parenthesis separating the statements!
df.filter((df['Close'] < 200) & (df['Open'] > 200)).show()
```

```python
df.filter((df['Close'] < 200) & ~(df['Open'] > 200)).show()
```

```python
# df.filter(df['Low'] == 197.16).show()
df.filter('Low =197.16').show()
```

```python
# Make sure to add in the parenthesis separating the statements!
df.filter( (df["Close"] < 200) | (df['Open'] > 200) ).show()
```

```python
# Collecting results as Python objects
result = df.filter('Low =197.16').collect()
result
```
